```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import luxpy as lx
import luxpy.toolboxes.spdbuild as spb 
```

```python
pump = lx.spdbuild.mono_led_spd(390, 20)
```

```python
# Set peak wavelengths of monochromatic leds:
peakwl = [420,460, 680]

# Set Full-Width-Half-Maxima of monochromatic leds:
fwhm = [60,100,50] 

# Set phosphor strengths:
strength_ph = [1., 0.4, 0]

# Set phosphor 1 parameters:
strength_ph1 = [1.8, 1.3, 0]
peakwl_ph1 = [520, 500, 1]
fwhm_ph1 = [200, 150, 60]

# Set phosphor 2 parameters:
strength_ph2 = [2, 1, 0]
peakwl_ph2 = [680, 630, 590]
fwhm_ph2 = [100, 100, 100]

S = spb.spd_builder(peakwl = peakwl, fwhm = fwhm,\
                    strength_ph = strength_ph, \
                    strength_ph1 = strength_ph1,\
                    peakwl_ph1 = peakwl_ph1,\
                    fwhm_ph1 = fwhm_ph1,\
                    strength_ph2 = strength_ph2,\
                    peakwl_ph2 = peakwl_ph2,\
                    fwhm_ph2 = fwhm_ph2
)
```

```python
plt.plot(S[0][30:], S[1][30:], label = 'Normalized Fluorophore Spectral Emission')
plt.plot(pump[0], pump[1], c = 'purple', alpha = 1, label = 'Normalized Modeled UV LED')
plt.ylabel('Normalized Spectral Power Distribition')
plt.xlabel('Wavelength (nm)')
plt.fill(pump[0], pump[1], c = 'purple', alpha = 0.5)
plt.fill_between(S[0][30:], S[1][30:], where=S[1][30:]>=[0]*len(S[1][30:]), interpolate = True, color = 'yellow', alpha = 0.5)
plt.grid()
plt.xlim(350, 780)
plt.legend(loc = (1.05, .8))
plt.show()
```
